### 1A

```python
def td_lambda_tabular(traces: Iterable[Iterable[mp.TransitionStep[S]]],
                      vf : Mapping[S, float], 
                      gamma: float,
                      lambd: float) -> Iterable[Mapping[S, float]]:
    
    for i  in traces:
        trace = {t : 0 for t in vf}
        for j in i:
            def possible(state, t):
                if state != t:
                    return 0
                return 1
            
            trace = {t : gamma * lambd * trace[t] + possible(j.state, t) for t in trace}
            error = j.reward + gamma * vf[j.next_state] - vf[j.state]
            # alpha = 0.02
            yield {s : vf[v] + 0.02 * error * trace[v] for v in vf}
```

### 1b

$$
G_t - V(S_t) = R_{t + 1} - V(S_t) + \gamma \cdot R_{t + 2} + \gamma^2 \cdot R_{t + 3} + \cdots\\
= ( R_{t + 1} + \gamma \cdot V(S_{t + 1}) - V(S_t) ) + ( \gamma \cdot R_{t + 2} + \gamma^2 \cdot V(S_{t + 2}) - \gamma \cdot V(S_{t + 1}) ) + \cdots\\
= ( R_{t + 1} + \gamma \cdot V(S_{t + 1}) - V(S_t) ) + \gamma \cdot ( R_{t + 2} + \gamma \cdot V(S_{t + 2}) - V(S_{t + 1}) ) + \cdots\\
G_t - V(S_t) = \sum_{u = t}^{T - 1} \gamma^{u - t} (R_{u + 1} + \gamma \cdot V(S_{u + 1}) - V(S_{u}) )\\
\text{QED}
$$

```python
mrp = SimpleInventoryMRPFinite(capacity = 2,
                               poisson_lambda = 1.,
                               holding_cost = 1.5,
                               stockout_cost = 10.)
```

```python
reward_traces = mrp.reward_traces(Choose(mrp.non_terminal_states))
vf = {s : 0 for s in mrp.non_terminal_states}

traces = 100
episodes = 1000

*traces, eval_vf = itertools.islice(
    td_lambda_tabular(traces = reward_traces,
                      vf = vf,
                      gamma = 0.9,
                      lambd = 0.5
                     )
    , traces * episodes)

#print(eval_vf)
```

```python
print("DP vs TD lambda")
mrp.display_value_function(gamma = 0.9)
pprint({s : round(eval_vf[s], 3) for s in mrp.non_terminal_states})
```

```python
errors: list

lambda_l = [x / 10.0 for x in range(0, 10, 1)]
for i, l in enumerate(lambda_l):
    td = td_lambda_tabular(traces = reward_traces,
                           vf = vf, 
                           gamma = 0.9,
                           lambd = l)
    
    errors.append(i[
            # a = 0, b = 0
            NonTerminal(state = InventoryState(on_hand = 0, on_order = 0))
            ]
        for i in list(itertools.islice(td, episodes)))

#print(errors)
```

```python
plt.xlabel("Steps")
plt.ylabel("Error")
for i, l in enumerate(lambda_l):
    plt.plot(errors[i],label='lambda='+str(l))
plt.legend()
plt.show()
```

### 1d


```python
class RandomWalkMRP2D(FiniteMarkovRewardProcess[Tuple[int, int]]):
    b1: int
    b2: int
    p_up: float
    # p_down = 1 - p_up
    p_left: float
    # p_right = 1 - p_left

    def __init__(
        self,
        b1: int,
        b2: int,
        p_up: float,
        p_left: float
    ):
        self.b1 = b1
        self.b2 = b2
        self.p_up = p_up
        self.p_down = 1 - p_up
        self.p_left = p_left
        self.p_right = 1 - p_left
        super().__init__(self.get_transition_map())

    def get_transition_map(self) -> \
            Mapping[Tuple[int, int], Categorical[Tuple[Tuple[int, int], float]]]:
        b1b2: List[Tuple[int, int]] = [(x, y) for x in range(1, self.b1) for y in range(1, self.b2)]
        d: Dict[Tuple[int, int], Categorical[Tuple[Tuple[int, int], float]]] = {
            (vert, horz) : Categorical({
                ((vert + 1, horz + 1), 0 if (self.b1 - 1 > vert) and (self.b2 - 1 > horz) else 1): self.p_up * self.p_left,
                ((vert - 1, horz + 1), 0 if (self.b2 > horz) else 1): self.p_down * self.p_left,
                ((vert + 1, horz - 1), 0 if (self.b1 > vert) else 1): self.p_up * self.p_right,
                ((vert - 1, horz - 1), 0): self.p_down * self.p_right,
            }) for vert, horz in b1b2
        }
        return d
```

```python
random_walk: RandomWalkMRP2D = RandomWalkMRP2D(
    b1 = 10,
    b2 = 10,
    p_up = 0.5,
    p_left = 0.5,
)
compare_td_and_mc(
    fmrp=random_walk,
    gamma=1.0,
    mc_episode_length_tol=1e-6,
    num_episodes=700,
    learning_rates=[(0.01, 1e8, 0.5), (0.05, 1e8, 0.5)],
    initial_vf_dict={s: 0.5 for s in random_walk.non_terminal_states},
    plot_batch=7,
    plot_start=0
)
```

### LSPI

```python
from rl.chapter8.optimal_exercise_bin_tree import OptimalExerciseBinTree
from rl.chapter12.optimal_exercise_rl import OptimalExerciseRL, fitted_lspi_put_option
from rl.gen_utils.plot_funcs import plot_list_of_curves


spot_price_val: float = 100.0
strike: float = 100.0
expiry_val: float = 1.0
rate_val: float = 0.05
vol_val: float = 0.25
num_steps_val: int = 200

opt_ex_bin_tree: OptimalExerciseBinTree = OptimalExerciseBinTree(
    spot_price=spot_price_val,
    payoff=lambda _, x: max(strike - x, 0),
    expiry=expiry_val,
    rate=rate_val,
    vol=vol_val,
    num_steps=num_steps_val
)

vf_seq, policy_seq = zip(*opt_ex_bin_tree.get_opt_vf_and_policy())
ex_boundary: Sequence[Tuple[float, float]] = \
    opt_ex_bin_tree.option_exercise_boundary(policy_seq, False)
time_pts, ex_bound_pts = zip(*ex_boundary)

label = ("Binomail Tree Put Option Exercise Boundary")
plot_list_of_curves(
    list_of_x_vals=[time_pts],
    list_of_y_vals=[ex_bound_pts],
    list_of_colors=["b"],
    list_of_curve_labels=[label],
    x_label="Time",
    y_label="Underlying Price",
    title=label
)

am_price: float = vf_seq[0][NonTerminal(0)]
print(f"Binomial Tree American Price = {am_price:.3f}")

dt: float = expiry_val / num_steps_val

num_training_paths: int = 5000
spot_price_frac_val: float = 0.0

lspi_training_iters: int = 8

split_val: int = 1000

num_scoring_paths: int = 10000

def payoff_func(_: float, s: float) -> float:
    return max(strike - s, 0.)

opt_ex_rl: OptimalExerciseRL = OptimalExerciseRL(
    spot_price=spot_price_val,
    payoff=payoff_func,
    expiry=expiry_val,
    rate=rate_val,
    vol=vol_val,
    num_steps=num_steps_val
)

european_price: float = opt_ex_rl.european_put_price(strike)
print(f"European Price = {european_price:.3f}")

training_data: Sequence[TrainingDataType] = opt_ex_rl.training_sim_data(
    num_paths=num_training_paths,
    spot_price_frac=spot_price_frac_val
)
print("Generated Training Data")

flspi: LinearFunctionApprox[Tuple[float, float]] = fitted_lspi_put_option(
    obj=opt_ex_rl,
    strike=strike,
    expiry=expiry_val,
    training_data=training_data,
    training_iters=lspi_training_iters,
    split=split_val
)

print("Fitted LSPI Model")

for step in [0, int(num_steps_val / 2), num_steps_val - 1]:
    prices: np.ndarray = np.arange(120.0)
    exer_curve: np.ndarray = opt_ex_rl.exercise_curve(
        step=step,
        prices=prices
    )
    cont_curve_lspi: np.ndarray = opt_ex_rl.continuation_curve(
        func=flspi,
        step=step,
        prices=prices
    )
    plt.plot(
        prices,
        cont_curve_lspi,
        "r",
        prices,
        exer_curve,
        "b"
    )
    time: float = step * expiry_val / num_steps_val
    plt.title(f"LSPI Curve for Time = {time:.3f}")
    plt.show()

ex_boundary_lspi: Sequence[float] = opt_ex_rl.put_option_exercise_boundary(
    func=flspi,
    strike=strike
)
time_pts: Sequence[float] = [i * dt for i in range(num_steps_val + 1)]
plt.plot(time_pts, ex_boundary_lspi, "r")
plt.title("LSPI Exercise Boundary")
plt.show()

scoring_data: np.ndarray = opt_ex_rl.scoring_sim_data(
    num_paths=num_scoring_paths
)

lspi_opt_price: float = opt_ex_rl.option_price(
    scoring_data=scoring_data,
    func=flspi
)
print(f"LSPI Option Price = {lspi_opt_price:.3f}")
```
